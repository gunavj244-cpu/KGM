# Cleaned Outliers Boxplot

```python
columns_to_clean = ['GDP', 'Alcohol', 'infant deaths', 'Adult Mortality','Population','Schooling',' BMI ','Total expenditure']

for col in columns_to_clean:
    Q1 = life[col].quantile(0.25)
    Q3 = life[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    life = life[(life[col] >= lower_bound) & (life[col] <= upper_bound)]
```

# Hypothesis test

# T-Test

```python
# Load and clean data
life = pd.read_csv("Life Expectancy1.csv")
life['Life expectancy '] = pd.to_numeric(life['Life expectancy '], errors='coerce')

# Group by status
developed = life[life['Status'] == 'Developed']['Life expectancy '].dropna()
developing = life[life['Status'] == 'Developing']['Life expectancy '].dropna()

# Perform t-test
t_stat, p_val = ttest_ind(developed, developing, equal_var=False)
print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
```

# U-Test



```python
developed_alcohol = life[life['Status'] == 'Developed']['Alcohol'].dropna()
developing_alcohol = life[life['Status'] == 'Developing']['Alcohol'].dropna()

u_stat, p_val = mannwhitneyu(developed_alcohol, developing_alcohol, alternative='two-sided')
print(f"U-statistic: {u_stat:.3f}, P-value: {p_val:.3f}")
```

# Anova Test

```python
from scipy.stats import f_oneway

group1 = life[life['Status'] == 'Developed']['Life expectancy '].dropna()
group2 = life[life['Status'] == 'Developing']['Life expectancy '].dropna()

stat, p = f_oneway(group1, group2)
print(f"ANOVA F-statistic: {stat:.3f}, p-value: {p:.5f}")
```
