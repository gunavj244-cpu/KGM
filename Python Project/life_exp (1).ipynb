# Hypothesis test

# T-Test

```python
from scipy.stats import ttest_ind

# Split data by status
developed = life[life['Status'] == 'Developed']['Life expectancy ']
developing = life[life['Status'] == 'Developing']['Life expectancy ']

# Run t-test
t_stat, p_val = ttest_ind(developed, developing, nan_policy='omit')

print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
```

- T-statistic = 10.413 → Large positive value, meaning the average life expectancy in developed countries is much higher than in developing countries.

- P-value = 0.000 → Extremely small, far less than 0.05.
⇒ There is a statistically significant difference in life expectancy between developed and developing countries.

- We separated the dataset into two groups:
>→ Countries with Developed status
>
>→ Countries with Developing status

- The Independent T-test compares the means of the two groups.

# T-Test



```python
from scipy import stats

developed = life[life['Status'] == 'Developed']['Schooling'].dropna()
developing = life[life['Status'] == 'Developing']['Schooling'].dropna()

t_stat, p_val = stats.ttest_ind(developed, developing, equal_var=False)
print("T-test (Schooling): t=%.3f, p=%.3f" % (t_stat, p_val))

if p_val < 0.05:
    print(" Null Hypothesis Rejected → Schooling is significantly different between Developed and Developing countries.")
else:
    print(" Fail to Reject Null Hypothesis → No significant difference in Schooling between Developed and Developing countries.")
```


- Developed countries generally have higher average years of schooling compared to developing countries.
    
- The T-test checks if this difference is statistically significant.

- This shows how education levels vary between developed and developing countries.



# Anova Test

```python
from scipy import stats
import pandas as pd

life['GDP_group'] = pd.qcut(life['GDP'], 3, labels=['Low', 'Medium', 'High'])

low = life[life['GDP_group'] == 'Low']['Life expectancy ']
medium = life[life['GDP_group'] == 'Medium']['Life expectancy ']
high = life[life['GDP_group'] == 'High']['Life expectancy ']

f_stat, p_val = stats.f_oneway(low.dropna(), medium.dropna(), high.dropna())
print("ANOVA: F=%.3f, p=%.3f" % (f_stat, p_val))

alpha = 0.05
if p_val < alpha:
    print(" Reject Null Hypothesis: Life Expectancy differs across GDP groups.")
else:
    print(" Fail to Reject Null Hypothesis: No significant difference in Life Expectancy across GDP groups.")
```

ANOVA checks whether average life expectancy is the same across all years.

- countries were divided into Low, Medium, High GDP based on their economic level.

- Low GDP countries → Life expectancy is generally low (people live fewer years).

- Medium GDP countries → Life expectancy is moderate

- High GDP countries → Life expectancy is usually high (people live longer).

- GDP is an important factor influencing how long people live.
